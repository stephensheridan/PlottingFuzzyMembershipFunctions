```python
import numpy as np
# Remove the following line after first run!
%pip install scikit-fuzzy
import skfuzzy as fuzz
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
# Example of setting up the x-vals for Fuzzy Risk example in the lecture notes.
# start, end, increment step
# Input
funding_range = np.arange(0, 100, 1)  
# Input
staffing_range = np.arange(0, 100, 1)  
# Output variable
risk_range = np.arange(0, 100, 1)
```

```python
# Create the project Funding variable and terms
funding_inadequate = fuzz.trapmf(funding_range,[0, 0, 20, 30])  
funding_marginal = fuzz.trimf(funding_range,[20,50,80])  
funding_adequate = fuzz.trapmf(funding_range,[60, 80, 100, 100])  
```

```python
# Plot the Funding variable and save it
fig, ax = plt.subplots(nrows=1, figsize=(10, 5))
ax.plot(funding_range, funding_inadequate, 'b', linewidth=1.5, label='Inadequate')  
ax.plot(funding_range, funding_marginal, 'g', linewidth=1.5, label='Marginal')  
ax.plot(funding_range, funding_adequate, 'r', linewidth=1.5, label='Adequate')  
ax.set_title('Project Funding')  
ax.legend()
plt.savefig("Funding.jpg")
```

```python
# Create the project Staffing variable and terms
staffing_small = fuzz.trapmf(staffing_range,[0, 0, 30, 57])  
staffing_large = fuzz.trapmf(funding_range,[40,60,100,100])  
```

```python
# Plot the Staffing variable and save it
fig, ax = plt.subplots(nrows=1, figsize=(10, 5))
ax.plot(staffing_range, staffing_small, 'b', linewidth=1.5, label='Small')  
ax.plot(staffing_range, staffing_large, 'g', linewidth=1.5, label='Large')  
ax.set_title('Project Staffing')  
ax.legend()
plt.savefig("Staffing.jpg")
```

```python
# Create the project Risk output variable and terms
risk_low = fuzz.trapmf(risk_range,[0, 0, 20, 40])  
risk_normal = fuzz.trimf(risk_range,[20,50,80])  
risk_high = fuzz.trapmf(risk_range,[60,80,100,100])  
```

```python
# Plot the Risk output variable and save it
fig, ax = plt.subplots(nrows=1, figsize=(10, 5))
ax.plot(risk_range, risk_low, 'b', linewidth=1.5, label='Low')  
ax.plot(risk_range, risk_normal, 'g', linewidth=1.5, label='Normal')  
ax.plot(risk_range, risk_high, 'r', linewidth=1.5, label='High')  
ax.set_title('Project Risk')  
ax.legend()
plt.savefig("Risk.jpg")
```
